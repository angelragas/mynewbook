```python
#To clean the rows in Year and Series columns (our selection)
filtered_data_cleaned = filtered_data.dropna(subset=["Year", "Series"])

#To select our main variable
filtered_data_total = filtered_data_cleaned[
   (filtered_data_cleaned["Series"] == "Population mid-year estimates (millions)")
]

filtered_data_total
```

```python
#Step 4B.A Compute the mean
mean_value = filtered_data_total['Value'].mean()

mean_value
```

```python
#Step 4B.B Compute the median
median_value = filtered_data_total['Value'].median()

median_value
```

```python
#Step 4B.C Compute the mode
filtered_data_total = filtered_data_total.dropna(subset=['Value'])

mode_value = filtered_data_total['Value'].mode()

print("Mode Value:", mode_value[0] if not mode_value.empty else "No mode")
```

```python
#Step 5 (hard way)

#To contain only the non-NaN values from the Value column in the filtered_data_total df, filtering out any NaN values
data_values = [value for value in filtered_data_total['Value'] if pd.notna(value)]
```

```python
#Compute the mean
mean_value = sum(data_values) / len(data_values)
mean_value
```

```python
#Compute the median
sorted_data = sorted(data_values)
n = len(sorted_data)

if n % 2 == 1:
    median_value = sorted_data[n // 2]
else:
    median_value = (sorted_data[n // 2 - 1] + sorted_data[n // 2]) / 2

median_value
```

```python
#Compute the mode
#Importing the multimode module
from statistics import multimode

mode_values = multimode(data_values)
mode_values
```

```python
counts = {}
for value in data_values:
    counts[value] = counts.get(value, 0) + 1

max_count = max(counts.values())
modes = [key for key, count in counts.items() if count == max_count]

print(f"Moda: {modes}")

#To be honest, Professor and TA, in this part a friend (engineer system) helped me, trying to understand the logic.
```

```python
#Step 6 (table)
#Population mean per year
average_by_year = filtered_data_total.groupby('Year')['Value'].mean().round(1)

#Creating the df
average_by_year_df = average_by_year.reset_index()
average_by_year_df.columns = ["Year", "Average Population"]

# Adding some format to the headers (only for visualization)
title = "Average population (in millions) per year"
print(f"{title:^50}")
print("=" * 50)
#The code center came from here https://www.w3schools.com/
print(f"{'Year':<10} {'Population (Mill.)':<20}")
print("-" * 50)

for _, row in average_by_year_df.iterrows():
    print(f"{int(row['Year']):<10} {row['Average Population']:>10}")

#I found 'iterrows' definition in https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.iterrows.html
```

```python
#Step 6 (sparklines)

average_by_year_df['Sparkline'] = average_by_year_df['Average Population'].apply(lambda x: '*' * round(x / 10))

# Adding some format to the headers (only for visualization)
title = "Average population (in millions) per year"
print(f"{title:^50}")
print("=" * 50)
#The code center came from here https://www.w3schools.com/
print(f"{'Year':<10} {'Population (Mill.)':<20}")
print("-" * 50)

for index, row in average_by_year_df.iterrows():
    print(f"{row['Year']:<10} {row['Sparkline']:<20}")
```
